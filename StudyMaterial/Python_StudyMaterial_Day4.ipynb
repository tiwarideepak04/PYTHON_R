```python
# Write a function to find the factorial, allow only positive integer alone

## 6! FACTORIAL is denoted by 1*2*3*4*5*6 = 720

#num = 6

num = int(input("Enter a number: "))

def fact(num):
    factorial = 1

    # check if the number is negative, positive or zero
    if num < 0:
       print("Sorry, factorial does not exist for negative numbers")
    elif num == 0:
       print("The factorial of 0 is 1")
    else:
       for i in range(1,num + 1):
           factorial = factorial*i
    print("The factorial of",num,"is",factorial)
    
    
#call the function fact

fact(num)
```

```python
# factorial using while loop
# 6! FACTORIAL is denoted by 1*2*3*4*5*6 = 720
num = int(input("enter a number: "))
 
def fact_while(num):
    fac = 1
    i = 1

    while i <= num:
        fac = fac * i
        i = i + 1

    print("factorial of ", num, " is ", fac)
    
fact_while(num)
```

###### global keyword we are telling that a is globally defined, so when we are changing a’s value inside the function it is actually changing for the a outside of the function also.

```python
# end to end example for local and gloabl variable inside function


a = 1
  
# Uses global because there is no local 'a' 

def f(): 
    print( 'Inside f() : ', a )
  
# Variable 'a' is redefined as a local 
def g():     
    a = 2 # modified the value of a
    print('Inside g() : ',a )
  
# Uses global keyword to modify global 'a' 
def h():     
    global a 
    a = 3
    print('Inside h() : ',a )
  
# Global scope 
print('global : ',a )
f() 
print('global : ',a )
g() 
print('global : ',a )
h() 
print('global : ',a )

print("value:",a)
```

# Difference between reduce and accumulator


```python
# Both reduce() and accumulate() can be used to calculate the summation of a sequence elements.
# reduce() is defined in “functools” module, accumulate() in “itertools” module.
# reduce() stores the intermediate result and only returns the final summation value. 
# accumulate() returns a list containing the intermediate results.The last number of the 
# list returned is summation value of the list.
```

```python
# importing itertools for accumulate() 
import itertools 
  
# importing functools for reduce() 
import functools 
```

```python
# initializing list  
lis = [ 1, 3, 4, 10, 4 ] 
```

```python
# priting summation using accumulate() 
print ("The summation of list using accumulate is :",end="") 
print (list(itertools.accumulate(lis,lambda x,y : x+y))) 
```

```python
# priting summation using reduce() 
print ("The summation of list using reduce is :",end="") 
print (functools.reduce(lambda x,y:x+y,lis))
```
